```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from pathlib import Path

%cd C:\Users\jacki\Desktop\Thesis\data\MeshRIR\MeshRIR
import irutilities as irutil
```

```python
#========== Load IR data ==========
%cd C:\Users\jacki\Desktop\Thesis\data\MeshRIR\S32-M441_npy
sessionName = "S32-M441_npy"  # "S32-M441_npy"
sessionPath = Path('.').joinpath(sessionName)
posMic, posSrc, ir = irutil.loadIR(sessionPath)

# Sampling rate
samplerate = 48000
```

```python
#========== Plot geometry ==========
irutil.drawGeometry(posSrc, posMic)
```

```python
#======== Normalize IR data =======
ir_export = np.array(ir)
max_amp = np.max(np.abs(ir_export))
ir_export = ir_export/max_amp
```

```python
#======== Normalize positional data =======
pos = np.concatenate((posMic,posSrc), axis=0)
posrange = np.nanmax(np.abs(pos), axis=0)
posMic_export = posMic/posrange
posSrc_export = posSrc/posrange
```

```python
#========== IR plots ==========
def plotIRBeginning(srcIdx, micIdx, si=5700, ei=6100):
    ir_plt = ir_export[srcIdx[:,np.newaxis], micIdx, si:ei]
    ir_plt = np.reshape(ir_plt, (-1,ir_plt.shape[-1]))
    t = np.linspace(si,ei,ei-si)/samplerate
    plt.plot(np.tile(t, (len(srcIdx)*len(micIdx),1)).T,ir_plt.T)
    plt.show()

micIdx = np.linspace(0,1, dtype=np.intp)
srcIdx = np.linspace(0,31, dtype=np.intp)
plotIRBeginning(srcIdx, micIdx)
```
