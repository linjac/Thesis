```python
# Training loss across models
import numpy as np
import matplotlib.pyplot as plt
```

```python
output_dir = './output/'
model_names = ['CNN1d_L1', 'CNN1d_UNET_L1', 'CNN1d_UNET_L2_2', 'CNN1d_UNET_L2_3']
```

```python
fig, ax = plt.subplots(1)  
for model in model_names:
    filename = output_dir + model + '_train_loss_L1.npy'
    ax.plot(np.load(open(filename, 'rb')), alpha=0.6, linewidth=1)
#ax.plot(loss_over_time)
#ax.plot(loss_over_time01)
#ax.plot(loss_over_time02)
#ax.plot(loss_over_time_mlp)
ax.legend(model_names)
ax.set_yscale('log')
ax.set_title(f'Training loss L1 across models')
#plt.ylim([7e-5, 1e-4])
plt.savefig("_compare_models_training_loss_L1.png", bbox_inches="tight")
plt.show()
```
