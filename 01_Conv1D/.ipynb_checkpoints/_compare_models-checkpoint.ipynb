```python
# Training loss across models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
output_dir = './output/'
model_names = ['CNN1d_L1', 
               'MLP0_L1',
               #'CNN1d_UNET_L1', 
               #'CNN1d_UNET_L2',
               'CNN1d_UNET_L2_2', 
               #'CNN1d_UNET_L2_3', 
               #'CNN1d_UNET2_L1', 
               #'CNN1d_UNET2_L2', 
               #'CNN1d_UNET3_L1',
               #'CNN1d_UNET4_L1',
               
               'CNN1d_UNET41_L1',
               #'CNN1d_UNET_BATCHED0_L1',
               'CNN1d_UNET40_L1',
               'CNN1d_UNET40_BATCHED_L1',
              ]
```

```python
plt.rcParams['figure.dpi'] = dpi = 150
plt.rcParams['savefig.dpi'] = dpi
plt.rcParams['figure.figsize'] = [15, 8]

fig, ax = plt.subplots(1)  
for model in model_names:
    filename = output_dir + model + '_train_loss_L1.npy'
    loss = np.load(open(filename, 'rb'))
    loss = pd.Series(loss).rolling(5).min().dropna().tolist()
    ax.plot(loss, alpha=0.6, linewidth=1)
#ax.plot(loss_over_time)
#ax.plot(loss_over_time01)
#ax.plot(loss_over_time02)
#ax.plot(loss_over_time_mlp)
ax.legend(model_names)
ax.set_yscale('log')
ax.set_title(f'Training loss L1 across models')
plt.ylim([1e-5, 1e-3])
#plt.xlim([50000, 60000])
plt.savefig("_compare_models_training_loss_L1.png", bbox_inches="tight")
plt.show()
```
